```python
import random

lorem = """Min læremester har sagt at en konge skal kunne tale 
Platin, for at kunne konversere med udsendinge fra Det 
Gromerske Rige... Jeg må hellere øve mig lidt...

"Lorem ipsum dolor sit amet, consectetur adipiscing elit, 
sed do eiusmod tempor incididunt ut labore et dolore magna 
aliqua. Ut enim ad minim veniam, quis nostrud exercitation 
ullamco laboris nisi ut aliquip ex ea commodo consequat. 
Duis aute irure dolor in reprehenderit in voluptate velit 
esse cillum dolore eu fugiat nulla pariatur. Excepteur 
sint occaecat cupidatat non proident, sunt in culpa qui 
officia deserunt mollit anim id est laborum."

I dag fik jeg {} til morgenmad. Det smagte {}.
"""

guldfisk = """I dag gav min læremester mig en guldfisk ved navn Kevin. Han sagde 
at den er bedre til at tale Platin end mig, så han mente at jeg måske 
kunne lære noget af den. Godt at guldfisken ikke kan sladre til læremesteren.
"""

drage = """I dag skete der endeligt noget sjovt. 
Jeg fik besøg af Filikodus! Han havde sin flotte drage med. 
Jeg kunne godt tænke mig en drage.
"""

with open('DAGBOG.txt.tmp', 'w') as fo:

    # 42 days of Latin træning
    for i in range(1, 365*10):
        fo.write(f'-------------\n\nDAG {i}\n\n')
        if i == 1042:
            fo.write(guldfisk)
        elif i == 2769:
            fo.write(drage)
        else:
            morgenmad = random.choice(['æg', 'ostemadder', 'havregrød', 'frugt'])
            smag = random.choice(['skidt', 'forfærdeligt', 'skønt', 'kedeligt'])
            fo.write(lorem.format(morgenmad, smag))
        fo.write(f'\n')
```

```python
import os

retninger = ['nord', 'syd', 'øst', 'vest']

rotte = """
En fedtet rotte sidder for enden af gangen.


    (\,/)
    oo   '''//,        _
  ,/_;~,        \,    / '
  "'   \    (    \    !
        ',|  \    |__.'
        '~  '~----''
"""

kat = """
En hvid kat sidder for enden af gangen. 
Der står Wozniak på dens navneskilt.

    /\__/\
   /`    '\
 === 0  0 ===
   \  --  /
  /        \
 /          \
|            |
 \  ||  ||  /
  \_oo__oo_/#######o
"""

def labyrint(sti, dybde):
    if dybde <= 0:
        yield '/'.join(sti)
    else:
        for ret in retninger:
            yield from labyrint(sti + [ret], dybde - 1)

for i, sti in enumerate(labyrint([], 4)):
    base = 'eventyr/vindebro/slot/trappe_ned/labyrint'
    sti = f'{base}/{sti}'
    os.makedirs(sti)
    dyr = 'KAT' if i == 41 else 'ROTTE'
    indhold = kat if i == 41 else rotte
    dyrloc = f'{sti}/{dyr}'
    with open(dyrloc, 'w') as fo:
        fo.write(indhold)
    
```
